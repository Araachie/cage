# Video composition demo with CAGE

We start with initializing a Demo instance and loading the frames and the models.

```python
demo = Demo("clevrer", source_image=None, device="cuda:0", steps=10, num_frames=9, past_horizon=4, random_time=False)
```

Once the demo is launched, you can perform the following steps to select and move objects:

1) Press "C" to enter the Select mode. -> "C" will be displayed in the upper-right corner of the frame.
2) Use the mouse to highlight the selected tiles.
3) Press "N" to switch to the next frame.
4) Repeat 1-3) to select objects from other frames.
5) Press "V" to enter the Composition mode. -> "V" will be displayed in the upper-right corner of the frame.
6) Use the mouse to place objects. Multiple instances of the same object are allowed.
7) Press "M" to enter the Move mode. -> "M" will be displayed in the upper-right corner of the frame.
8) Use the mouse to drag the last selected region to a new location in the first future frame.
9) Repeat 8) to drag the last selected placed object to new locations in the second (third etc.) future frames.
10) Press "N" to switch to the next object.
11) Repeat 6-10) to place and move other objects.
12) Press "Enter" to generate the video. -> "F" will be displayed, when the video is ready.

```python
%matplotlib widget
demo.reinit()
demo.run(indices=[0, 1, 2])
```

Now we can play the genereated video.

```python
%matplotlib inline
_ = show_video(demo.generated_frames)
```
