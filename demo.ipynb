# Image animation demo with CAGE

We start with initializing a Demo instance and loading the frames and the models.

```python
demo = Demo("clevrer", device="cuda:0", num_frames=12, random_time=False)
```

Once the demo is launched, you can perform the following steps to select and move objects:

1) Press "C" to enter the Select mode. -> "C" will be displayed in the upper-right corner of the frame.
2) Use the mouse to highlight the selected tiles.
3) Press "M" to enter the Move mode. -> "M" will be displayed in the upper-right corner of the frame.
4) Use the mouse to drag the last selected region to a new location in the first future frame.
5) Repeat 4) to drag the last selected region to new locations in the second (third etc.) future frames.
6) Repeat 1-5) for other objects.
7) Press "Enter" to generate the video. -> "F" will be displayed, when the video is ready.

```python
%matplotlib widget
demo.reinit()
demo.run(index=47)
```

Now we can play the genereated video.

```python
%matplotlib inline
_ = show_video(demo.generated_frames)
```
